A collection of useful functions for generic use 

```python
# breakdown voltages
import numpy as np 

                # Kpp.cm^-1       SEY          Kpascal      cm 
def Vbr_get(A = 112.5, B =2737.5, gam = 0.95, p = 101, d = 1):
    return (B*p*d)/(np.log(A*p*d) - np.log(np.log(1+(1/gam))))

press= np.logspace(2,-3,20)
vr = Vbr_get(d= 10, gam=0.85, p=press)
print(vr)
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
plt.plot(press,vr)
ax =plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
plt.xlabel('Pressure [Pa]')
plt.ylabel('V_breakdown [V]')
plt.grid()
```
