# Covert Transport to human 
The PSI TRANSPORT code is a punch card style interface and very hard to look at, here is a code to take a input file and covert it into a human readable interface. Later versions are planed to convert the inptu into other languages too

```python
import os 
#input file 
#inputDir =  "/mnt/c/Users/stu_w/OneDrive/Software/TRANSPORT/scripts"
inputDir =  "/mnt/c/Users/warren_s/OneDrive/Software/TRANSPORT/scripts"
#filename =  "2MeV-AWE-4.dat"
#filename =  "55-a-7.dat"
filename =  "220-ARI1-10.dat"

FnDir =os.path.join(inputDir, filename)
#open file and copy to memory 
print("opening: ", FnDir)
with open(FnDir,'r') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
print("File has ", len(lines), " lines in mem")
      
```

# parsing the file to readable units
From the manual http://stefano.sinigardi.it/published/Manuale_Transport.pdf the inputs will be rewritten and comments removed 

```python
# is a number function 
def isAnumber(string = ''):
    
    if(string.find('-')>=0):
        string=string[string.find('-')+1:len(string)]
        #print(string)
    return string.replace('.','',1).isdigit()
#test 
#print(isAnumber("22.34sdfdsdf"))
```

```python
import numpy as np


def LookUpEl(El=0, com = ''):
    #elemet number 
    var=[]
    #has lable 
    c = com.split(' ')
    num = np.floor(float(c[0]))
    if(len(c)>1):        
        c = c[1:len(c)]
    else: 
        num = 0    
    # remove empty results
    c = [ele for ele in c if ele.strip()]
    #print(c)

    #process     
    if(int(num) == 0 ):# command for program 
        return {'EntryNo':El,'Element':'ProgramC', 'command': c[0]}
    #Beam init 
    if(int(num) == 1 ):
        dic = {'EntryNo':El,'Element':'Beam'}
        v = [ 'x[cm]', 'Px[mr]','y','Py','Blen[cm]','disp[%]','Pmom','','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
    
    if(num == 2 ):
        dic = {'EntryNo':El,'Element':'Rotation'}
        v = [ 'angle[deg]', '','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic        
    if(num == 3 ):
        dic = {'EntryNo':El,'Element':'Drift'}
        v = [ 'Len[m]', '','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 4):
        dic = {'EntryNo':El,'Element':'BendingMag'}
        v = [ 'Len[m]', 'angle[deg]','FeildGrad[n]','','']
        #print (c)
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        try: 
            th = dic['angle[deg]']
            l = dic['Len[m]']
            dic['BendRad'] = l/((th/180.)*np.pi)
        except:
            dic['BendRad'] =0            
        return dic
    if(num == 5):
        dic = {'EntryNo':El,'Element':'QP'}
        v = [ 'Len[m]', 'Feild[kG]','HalfApature','','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic        
    if(num == 6):
        dic = {'EntryNo':El,'Element':'Transform'}
        v = [ 'Len[m]', 'Feild[kG]','FeildGrad','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 7):
        dic = {'EntryNo':El,'Element':'BeamCentShift'}
        v = [ 'x[cm]', 'Px[mrad]','y[cm]', 'Py[mrad]','len[cm]','disp[%]','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 8):
        dic = {'EntryNo':El,'Element':'AlignTollerance'}
        v = [ 'x[cm]', 'Px[mrad]','y[cm]', 'Py[mrad]','z[cm]','rot[mrad]','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
       
    if(num == 9):
        dic = {'EntryNo':El,'Element':'RepeatContrl'}
        v = [ 'Nrepeat','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 10 or num == -10):
        dic = {'EntryNo':El,'Element':'FitConstr'}
        v = [ 'function', 'J','desiredVal]','accurcy','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 11):
        dic = {'EntryNo':El,'Element':'Accelerator'}
        v = [ 'Len[m]', 'EnergyGain[GeV]','PhaseLag[deg]','WaveLength','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        return 'Accelerator'
    if(num == 12):
        dic = {'EntryNo':El,'Element':'BeamRotElipse'}
        v = [ 'R12','R13','R14','R15','R16','R23','R24','R25','R26','R34','R35','R36', 'R45','R46','R56','','' ]
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 13 ):
        dic = {'EntryNo':El,'Element':'In/OutOpts'}
        v = [ 'ControlNum','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
       
    if(num == 14):
        dic = {'EntryNo':El,'Element':'RMatrix'}
        v = [ 'Rji', 'Rj2','Rj3','Rj4','Rj5','Rj6','j','']
        #print(c)
        for i, val in enumerate(c):
            #print(i, " : ", val, " : ", v[i])
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 15):
        dic = {'EntryNo':El,'Element':'Units'}
        v = [ 'code', 'Unit','ScaleFctr','']
        unit = ['','XY[cm]','PxPy[mr]','Y[cm]','Py[mr]','BunchLen[cm]','dP[%]','angle[deg]','Len[m]','Bfield[kG]','Mass[Me]','Pgain[GeV/c]', '']
        print(c)
        try: 
            opt  = int(float(c[0]))
            name = c[1]            
            val  = float(c[2])
            dic['oldUnit']  = unit[opt]
            dic['code']     = opt
            dic['newUnit']  = name[name.find('/')+1:-1]
            dic['ScaleFctr']= val
            
        except:
            dic["error"]  = 1           
        return dic
    
    if(num == 16):
        dic = {'EntryNo':El,'Element':'BendingMagPar'}
        v = [ 'QuadTermBend','P[cm]','M/m' ,'W/2[cm]',
             'g/2[cm]','Len[cm]','k1-fringe','k2-fringe',
             'Crv_ben_en_Face','Crv_ben_ex_Face',
             'FocPlaneRot','','','']
        try: 
            opt = int(float(c[0]))
            val = float(c[1])
            name= c[2]
            dic[v[opt]]= val
            dic['name'] = name[name.find('/')+1:-1]
        except:
            dic["error"] = 1           
        return dic
    
    if(num == 17):
        dic = {'EntryNo':El,'Element':'BendingMag'}
        v = [ '', '','','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 18):
        dic = {'EntryNo':El,'Element':'Sextupole'}
        v = [ 'Len[m]', 'Feild[kG]','','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 19):
        dic = {'EntryNo':El,'Element':'Solenoid'}
        v = [ 'Len[m]', 'Feild[kG]','FeildGrad','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 20):
        dic = {'EntryNo':El,'Element':'BeamRot'}
        v = [ 'Rot[deg]', '','','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
        
    if(num == 21):
        dic = {'EntryNo':El,'Element':'StrayFeild'}
        v = [ 'Len[m]', 'Feild[kG]','FeildGrad','']
        for i, val in enumerate(c):
            if(isAnumber(val)):            
                dic[v[i]]=float(val)
            else:
                dic['name'] = val[val.find('/')+1:-1]
        return dic
import numpy as np
# tor remove trailing numbers 
def round_sig(x, sig=2):
    try:
        if(x!=0):
            var = np.round(x, sig-int(np.floor(np.log10(np.abs(x))))-1)   
        else:
            var = x 
    except:
        var = x
    return var
    

```

```python
# strip the comments from the file 
cmds = []
for a in lines:
    a=a.lstrip()
    if(len(a)>0):        
        #remove white spaces 
        a.join(a.split())
        if ( a[0] == '/' or a[0] == '('or a[0] == ' ' or a[0] == '\n'or len(a)==0):
            #its a commnet line         
            #print("comment: ", a)
            a=''
        else:
            # strip comment
            cmtc = a.find('(')
            cmts = a.find('/*')
            cmds.append(a)
# now process commands
ElType = []
Zpos = 0
for i, c in enumerate(cmds):   
    #remove the ;
    if(c.find(';')>0):
        c = c[0:c.find(';')]
    #print("command: ", c)
    if(c.find(' ')>0):
        El = c[0:c.find(' ')]
    else:
        El = c
    #print("Element: ", El)
    if(isAnumber(El)):
        ThisEl = LookUpEl(El=i,com=c)
        if ('Len[m]' in ThisEl):
            Zpos += ThisEl['Len[m]']
            
            #print(Zpos)
        ThisEl['Zpos']=Zpos
        ElType.append(ThisEl)
       # print (ElType)
        
    
print(len(ElType), " commands loaded into the system")  
```

## now we have the data in we can rewrite it to human language in a file 

```python
fout = filename[0:filename.find('.')] + "_converted.txt"
with open (os.path.join(inputDir, fout), 'w') as f:
    #parse the lines
    f.write("element # \t type\t name \t properties\n")
    for dic in ElType:
        #print(dic)
        f.write(str(dic['EntryNo']))
        f.write('\t')
        f.write(str(dic['Element']))
        f.write('\t')
        try:
            f.write(str(dic['name']))            
        except:
            f.write(" ")          
        f.write('\t')
        for key in dic:
            if(key!= 'EntryNo' and key!= 'Element'and key!='name'):
                f.write("\t"+key + "\t " + str(dic[key]))
        f.write('\n')
print("file: " ,fout , " written to folder" , inputDir)
                
```

# elemets sum and properties 
Here the number of elements and there properties are summed up 

```python
#go over elemets 
Magarr =[]
Qparr = []
for ThisEL in ElType:
    #print("This ThisEL:\t", ThisEL)
    for key in ThisEL:
        #print("This Key:\t", key)
        # get sector magnets
        
        if(ThisEL[key]=='BendingMag' ):
            if (len(Magarr)>0):
                Magarr.append(ThisEL)
            else:
                Magarr = [ThisEL]
            
        #get QP's 
        if(ThisEL[key]=='QP'):
            if (len(Qparr)>0):
                Qparr.append(ThisEL)
            else:
                Qparr = [ThisEL]


fout = filename[0:filename.find('.')] + "_BLE_only.CVS"
fw = open(os.path.join(inputDir, fout), 'w')
                
print ("\n--------------------------------------\n\t\tMagnet's\n--------------------------------------\n")
fw.write("\n--------------------------------------\n\t\tMagnet's\n--------------------------------------\n")
for m in Magarr: 
    
    try:
        z= round_sig(m['Zpos'], sig =3)
    except:
        z = ''
    
    try:
        name= m['name']
    except:
        name = ''
    try:
        l= round_sig(m['Len[m]'], sig =3)
    except:
        l = ''
    try:
        Ha= round_sig(m['BendRad'], sig =3)
    except:
        Ha = ''
    try:
        f= round_sig(m['angle[deg]'], sig =3)
    except:
        f = ''    
    print('S[m]: ',z ,'\tname : ',name,'\tBendRad : ', Ha,'\tangle[deg] : ',f,'\tLen[m] : ',l)
    fw.write('S[m]: '+str(z) +'\tname : '+str(name)+'\tBendRad : '+str(Ha)+
            '\tangle[deg] : '+str(f)+'\tLen[m] : '+str(l)+"\n")
    
print ("\n-----------------------------------\n\t\tQP's\n-----------------------------------\n")
fw.write("\n-----------------------------------\n\t\tQP's\n-----------------------------------\n")
for q in Qparr: 
    try:
        z= round_sig(q['Zpos'], sig =3)
    except:
        z = ''
    try:
        name= q['name']
    except:
        name = ''
    try:
        l= round_sig(q['Len[m]'], sig =3)
    except:
        l = ''
    try:
        Ha= q['HalfApature']
    except:
        Ha = ''
    try:
        f= round_sig(q['Feild[kG]'], sig =3)
    except:
        f = ''
    print('S[m]: ',z ,'\tname : ',name,'\tLen[m] : ',l,'\tHalfApature : ',Ha,'\tFeild[kG] : ',f)   
    fw.write('S[m]: '+str(z) +'\tname : '+str(name)+'\tLen[m] : '+str(l)+
            '\tHalfApature : '+str(Ha)+'\tFeild[kG] : '+str(f)+"\n")

fw.close()  

```
